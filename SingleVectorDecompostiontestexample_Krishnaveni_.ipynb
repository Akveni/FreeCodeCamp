```python
import os 
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = Image.open('test.jpg')
print('Size(dimension)',img.size)
plt.imshow(img)
```

```python
imggray = img.convert('LA')
print(imggray.size)
imgmat = np.array(list(imggray.getdata(band=0)))
imgmat.shape = (imggray.size[1],imggray.size[0])
imgmat = np.matrix(imgmat)
plt.figure()
plt.imshow(imgmat,cmap='gray')

```

```python
U, S, V = np.linalg.svd(imgmat)
for r in range(5,70,20):
    mat = np.matrix(U[:,:r])@np.diag(S[:r])@np.matrix(V[:r,:])
    print('Size(dimension):',mat.size)
    plt.imshow(mat,cmap='gray')
    plt.title(f'Image after = {r}')
    plt.show()
    image1 = Image.fromarray((mat).astype(np.uint8))
image1.save('compressed.jpg')     
        
```

```python
print("size", image1.size)
```

```python
result=image1.convert('RGB')
```

```python
plt.figure()
plt.imshow(result, cmap = 'cool')
plt.title("Image after converting it into the gray scale pattern")
plt.show()
```
